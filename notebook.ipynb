# **College Placement Records Analytics Workflow**

```python
import pandas as pd
import numpy as np
df = pd.read_csv('Placement_Data_Full_Class.csv')
```

Display the first 5 rows

```python
print(df.head())
```

Check missing values for every column

```python
print(df.isnull().sum())
```

Handle missing values appropriately and explain your choice using comments

```python
if 'salary' in df.columns:
    df['salary'] = df['salary'].fillna(0)
    print("\n[INFO] Missing values in 'salary' have been filled with 0 (indicating unplaced students).")
```

Remove duplicate records

```python
duplicate_count = df.duplicated().sum()
df.drop_duplicates(inplace=True)
print(f"\n[INFO] Removed {duplicate_count} duplicate rows.")
```

Clean all text columns (`strip()`, `title()`, `lower()` where appropriate)

```python
text_columns = df.select_dtypes(include=['object']).columns
for col in text_columns:
    df[col] = df[col].astype(str).str.strip().str.title()
print("\n[INFO] Text columns successfully stripped and formatted to Title Case.")
```

Fix incorrect data types

```python
if 'salary' in df.columns:
    df['salary'] = df['salary'].astype(int)
```

Rename columns where necessary for consistency

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('.', '_', regex=False)
print("\n--- Cleaned Column Names ---")
print(df.columns.tolist())
```

Create at least **one new useful column** using Python

```python
degree_col = [col for col in df.columns if 'degree' in col and '_p' in col]
if degree_col:
    p_col = degree_col[0]
    df['performance_tier'] = pd.cut(df[p_col],
                                    bins=[0, 60, 75, 100],
                                    labels=['Average', 'First Class', 'Distinction'])
    print(f"\n[INFO] New feature 'performance_tier' created based on {p_col}.")
else:
    df['performance_tier'] = 'Standard'
```

Save the cleaned dataset as **cleaned_dataset.csv**

```python
df.to_csv('cleaned_dataset.csv', index=False)
print("\n[SUCCESS] Cleaned dataset saved as 'cleaned_dataset.csv'")
```
